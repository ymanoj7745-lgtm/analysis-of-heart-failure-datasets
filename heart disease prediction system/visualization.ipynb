```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

```

```python
df = pd.read_csv("Heart_Disease_Prediction.csv")
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["BP"])
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age vs Blood Pressure")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(df["Age"], df["BP"])
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age vs Blood Pressure")
plt.show()
```

```python
#as we can see 50 to 60 age grouped are high blood pressure
```

```python
plt.figure(figsize=(6,4))
plt.pie(df["Exercise angina"], df["Thallium"])
plt.xlabel("Exercise angina")
plt.ylabel("Thallium")
plt.title("Excercise angina vs Thanlium")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.stackplot(df["Exercise angina"], df["Thallium"])
plt.xlabel("Exercise angina")
plt.ylabel("Thallium")
plt.title("Excercise angina vs Thanlium")
plt.grid(True)
plt.show()
```

```python
#visulization using seaborn
plt.figure(figsize=(6,4))
sns.lineplot(df["Exercise angina"], df["Thallium"])
plt.xlabel("Exercise angina")
plt.ylabel("Thallium")
plt.title("Excercise angina vs Thanlium")
plt.grid(True)
plt.show()
```

```python
#visulization using seaborn
plt.figure(figsize=(6,4))
sns.scatterplot(df["Exercise angina"], df["Thallium"])
plt.xlabel("Exercise angina")
plt.ylabel("Thallium")
plt.title("Excercise angina vs Thanlium")
plt.grid(True)
plt.show()
```

```python
#visulization using seaborn
plt.figure(figsize=(6,4))
sns.boxplot(df["Exercise angina"], df["Thallium"])
plt.xlabel("Exercise angina")
plt.ylabel("Thallium")
plt.title("Excercise angina vs Thanlium")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(df["Age"], df["BP"])
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age vs Blood Pressure")
plt.show()
```

```python
plt.figure(figsize=(4,3))

corr = df[["Age", "BP"]].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="RdBu_r",
    vmin=-1,
    vmax=1
)

plt.title("Age vs Blood Pressure Correlation")
plt.show()
```

```python
sns.jointplot(
    data=df,
    x="Age",
    y="BP",
    kind="hex",
    cmap="viridis"
)
plt.show()
```
